```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items2 = purchase_data["Item ID"].nunique()

avg_price2 = purchase_data["Price"].mean().round(2)

total_purchases2 = purchase_data["Purchase ID"].count()

total_revenue2 = purchase_data["Price"].sum()


pa_df = pd.DataFrame(
            {"Unique Items": [unique_items2],
            "Average Price": [avg_price2],
            "Total Number of Purchases": [total_purchases2], 
            "Total Revenue": [total_revenue2]})
pa_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gen_spec = purchase_data.groupby("Gender")["SN"].nunique()

gen_df = pd.DataFrame(gen_spec)

pct = gen_spec.reset_index()['SN'].sum()
gen_pct =  gen_spec.groupby(level=0)\
    .apply(lambda x:100*x/pct)\
    .reset_index(name='Percentage of Player')\
    .set_index("Gender").round(2).astype(str) + '%'

gen_df["Percentage of Players"] = gen_pct["Percentage of Player"]

gen_df.rename(columns={"SN": "Total Count"}).sort_values(by="Total Count", ascending=False)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 99]
bins2 = pd.cut(purchase_data["Age"], bins, labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"], right=False)

pa_a_age = purchase_data["Age"]
pa_a_price = purchase_data["Price"]
pa_a_sn = purchase_data["SN"]

pa_df1 = pd.DataFrame()
pa_df1["Age"] = pa_a_age
pa_df1["Price"] = pa_a_price
pa_df1["SN"] = pa_a_sn
pa_df1["Age Range"] = bins2

pag_age_count = pa_df1.groupby("Age Range")["SN"].count()
pag_age_total = pa_df1.groupby("Age Range")["Price"].sum()
pag_age_avgpp = (pag_age_total/pag_age_count).round(2)

pa_df2 = pd.DataFrame()
pa_df2["Purchase Count"]=pag_age_count
pa_df2["Average Purchase Price"]=pag_age_avgpp
pa_df2["Total Purchase Value"]=pag_age_total
pa_df2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_name = purchase_data["SN"]
item_price = purchase_data["Price"]
ts1_df = pd.DataFrame()
ts1_df["SN"] = sn_name
ts1_df["Price"] = item_price

sn_count = ts1_df.groupby("SN")["Price"].count()
sn_item_price = ts1_df.groupby("SN")["Price"].unique()
total_item_sum = ts1_df.groupby("SN")["Price"].sum().round(2)
sn_avgpp = (total_item_sum/sn_count).round(2)

ts2_df = pd.DataFrame()
ts2_df['Purchase Count'] = sn_count
ts2_df["Average Purchase Price"] = sn_avgpp
ts2_df['Total Purchase Value'] = total_item_sum
ts2_df.sort_values(by="Purchase Count", ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_id = purchase_data["Item ID"]
item_name = purchase_data["Item Name"]
item_price = purchase_data["Price"]
pi_df = pd.DataFrame()
pi_df["Item ID"] = item_id
pi_df["Item Name"] = item_name
pi_df["Price"] = item_price

id_count = pi_df.groupby(["Item ID", "Item Name"])["Item Name"].count()
mp_item_price = pi_df.groupby(['Item ID', "Item Name"])["Price"].unique()
total_item_sum = pi_df.groupby(["Item ID", "Item Name"])["Price"].sum().round(2)

mp_df = pd.DataFrame()
mp_df['Purchase Count'] = id_count
mp_df["Price"] = mp_item_price
mp_df['Total Purchase Value'] = total_item_sum
mp_df.sort_values(by="Purchase Count", ascending=False)
```
